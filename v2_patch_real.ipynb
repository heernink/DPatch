```python
######## 패치 불러오기 ########
patch = Image.open('/Data2/hm22/DPatch/trained_patch/14/5011.png')

######## yolo 불러오기 ########
from yolov2 import Yolov2
import os

pretrained_model = os.path.join('/Data1/sm21/transform_based_defense/YOLOv2/dataset/data/pretrained', 'darknet19_448.weights')

patch_np = np.array(patch)
patch_np.shape
patch_np = patch_np.transpose((2,0,1))
patch_np.shape
patch_t = torch.from_numpy(patch_np)



####### defense 코드 ########
def test_dpatch_loc_masking():
    args = parse_args()
    args.conf_thresh = 0.005
    args.nms_thresh = 0.45
    if args.vis:
        args.conf_thresh = 0.5
    print('Called with args:')
    print(args)

    # prepare dataset

#     if args.dataset == 'voc07trainval':
#         args.imdbval_name = 'voc_2007_trainval'

#     elif args.dataset == 'voc07test':
#         args.imdbval_name = 'voc_2007_test'

#     else:
#         raise NotImplementedError

#     val_imdb = get_imdb(args.imdbval_name)

#     val_dataset = RoiDataset(val_imdb, train=False)
#     val_dataloader = DataLoader(val_dataset, batch_size=args.batch_size, shuffle=False)
       
    val_imdb = get_imdb('voc_2007_test')
    val_dataset = RoiDataset(val_imdb, train=False)
    val_dataloader = DataLoader(val_dataset, batch_size=args.batch_size, shuffle=False)
   
    # load model
    pretrained_model = os.path.join('/Data1/sm21/transform_based_defense/YOLOv2/dataset/data/pretrained', 'darknet19_448.weights')
    model = Yolov2(weights_file=pretrained_model)
#     model = Yolov2()
#     weight_loader = WeightLoader()
#     weight_loader.load(model, pretrained_model)
    # print('loaded')

    model_path = os.path.join(args.output_dir, args.model_name+'.pth')
    print('loading model from {}'.format(model_path))
    if torch.cuda.is_available():
        checkpoint = torch.load(model_path)
    else:
        checkpoint = torch.load(model_path, map_location='cpu')
    model.load_state_dict(checkpoint['model'])

    if args.use_cuda:
        model.cuda()

    model.eval()
    print('model loaded')

    dataset_size = len(val_imdb.image_index)

    all_boxes = [[[] for _ in range(dataset_size)] for _ in range(val_imdb.num_classes)]

    det_file = os.path.join(args.output_dir, 'detections.pkl')

    img_id = -1
    with torch.no_grad():
        for batch, (im_data, im_infos) in enumerate(val_dataloader):
            if args.use_cuda:
#                 print(im_data.shape)
                im_data_variable = Variable(im_data).cuda()
            else:
#                 print(im_data.shape)
                s1, s2 = patch_t.shape[2], patch_t.shape[2]
#                 print(s1,s2)
                im_data[:,:,0:s1,0:s2] = patch_t
#                 print(im_data.shape)
#                 print(im_data.dtype)
               
   
                window_size = 3
                stride = 3
#                 print(im_data)
                ker_var_list = []
                for i in range(0,im_data.shape[3],stride):
                    for j in range(0,im_data.shape[2],stride):
#                         print(j,i)
                        ker = im_data[:,:,j:min(im_data.shape[2],j+window_size),i:min(im_data.shape[3],i+window_size)]
#                         print(ker)
#                         print(torch.var_mean(ker)[0])
#                         print(ker)
                        ker_var_list.append(torch.var_mean(ker)[0])
#                 print(ker_var_list)
                ker_var_np = np.array(ker_var_list)
                limit_var_lower = ker_var_np.mean()- (ker_var_np.std()*3)
                limit_var_upper = ker_var_np.mean()+ (ker_var_np.std()*3)
#                 print(limit_var_upper)
               
               
                wit = torch.ones(im_data.shape)

                x_coor = []
                y_coor = []
#                 print(im_data.shape)
                for i in range(0,im_data.shape[3],stride):
                    for j in range(0,im_data.shape[2],stride):
                        ker = im_data[:,:,j:min(im_data.shape[2],j+window_size),i:min(im_data.shape[3],i+window_size)]
                        ker_var = torch.var_mean(ker)[0]
#                         print(limit_var_upper)
#         print(j,i)
                        if ker_var > limit_var_upper:
#                             print(limit_var_upper)
                            x_coor.append(j)
                            y_coor.append(i)
                            wit[j:min(im_data.shape[2],j+window_size),i:min(im_data.shape[3],i+window_size)]=0
                        if ker_var < limit_var_lower:
                            x_coor.append(j)
                            y_coor.append(i)
                            wit[j:min(im_data.shape[2],j+window_size),i:min(im_data.shape[3],i+window_size)]=0
#                 print(x_coor)            
                coor = pd.DataFrame({'x_coor':x_coor, 'y_coor':y_coor})
                loc_data = StandardScaler().fit_transform(coor[['x_coor','y_coor']])
                clustering = DBSCAN(eps=0.5, min_samples=5).fit(loc_data)
#                 print(len(clustering.labels_))
                coor['clust'] = clustering.labels_
                max_idx = np.argmax(coor['clust'].value_counts())
#                 print(max_idx)
                coor_m = coor[coor["clust"]==max_idx]
               
                x_min = min(coor_m["x_coor"])
                y_min = min(coor_m["y_coor"])
                x_max = max(coor_m["x_coor"])+window_size
                y_max = max(coor_m["y_coor"])+window_size
               
               
                im_data[:,:,x_min:x_max, y_min:y_max] = 0
               
                im_data_variable = Variable(im_data)

            yolo_outputs = model(im_data_variable)
            for i in range(im_data.size(0)):
                img_id += 1
                output = [item[i].data for item in yolo_outputs]
                im_info = {'width': im_infos[i][0], 'height': im_infos[i][1]}
                detections = yolo_eval(output, im_info, conf_threshold=args.conf_thresh,
                                       nms_threshold=args.nms_thresh)
                print('im detect [{}/{}]'.format(img_id+1, len(val_dataset)))
                if len(detections) > 0:
                    for cls in range(val_imdb.num_classes):
                        inds = torch.nonzero(detections[:, -1] == cls).view(-1)
                        if inds.numel() > 0:
                            cls_det = torch.zeros((inds.numel(), 5))
                            cls_det[:, :4] = detections[inds, :4]
                            cls_det[:, 4] = detections[inds, 4] * detections[inds, 5]
                            all_boxes[cls][img_id] = cls_det.cpu().numpy()

                if args.vis:
                    img = Image.open(val_imdb.image_path_at(img_id))
                    if len(detections) == 0:
                        continue
                    det_boxes = detections[:, :5].cpu().numpy()
                    det_classes = detections[:, -1].long().cpu().numpy()
                    im2show = draw_detection_boxes(img, det_boxes, det_classes, class_names=val_imdb.classes)
                    plt.figure()
                    plt.imshow(im2show)
                    plt.show()

    with open(det_file, 'wb') as f:
        pickle.dump(all_boxes, f, pickle.HIGHEST_PROTOCOL)

    val_imdb.evaluate_detections(all_boxes, output_dir=args.output_dir)





######## defense 코드 실행 ########
import argparse
from torch.autograd import Variable
from dataset.factory import get_imdb
from torch.utils.data import DataLoader
from dataset.roidb import RoiDataset
from util.network import WeightLoader
from yolo_eval import yolo_eval
import torch
import pickle

def parse_args():

    parser = argparse.ArgumentParser('Yolo v2')
    parser.add_argument('--dataset', dest='dataset',
                        default='voc07test', type=str)
    parser.add_argument('--output_dir', dest='output_dir',
                        default='output', type=str)
    parser.add_argument('--model_name', dest='model_name',
                        default='yolov2_epoch_20', type=str)
    parser.add_argument('--nw', dest='num_workers',
                        help='number of workers to load training data',
                        default=0, type=int)
    parser.add_argument('--bs', dest='batch_size',
                        default=1, type=int)
    parser.add_argument('--cuda', dest='use_cuda',
                        default=False, type=bool)
    parser.add_argument('--vis', dest='vis',
                        default=False, type=bool)

    args = parser.parse_args([])
    return args



start = torch.cuda.Event(enable_timing=True)
end = torch.cuda.Event(enable_timing=True)

start.record()
test_dpatch_loc_masking()
end.record()

# Waits for everything to finish running
torch.cuda.synchronize()

print(start.elapsed_time(end))
```
